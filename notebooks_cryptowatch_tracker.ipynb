```python
import requests
import time

# Instead of a database file, we use a Python list (In-Memory Storage)
history = []

print("Starting Live Tracker (In-Memory Mode)...")

for i in range(5):
    try:
        # 1. Fetch live data
        price = float(requests.get("https://api.coinbase.com/v2/prices/BTC-USD/spot").json()['data']['amount'])
        mood = requests.get("https://api.alternative.me/fng/").json()['data'][0]['value_classification']
        
        # 2. Store it in our list
        data_point = {"check": i+1, "price": price, "mood": mood}
        history.append(data_point)
        
        print(f"Captured: ${price} | Mood: {mood}")
        time.sleep(3)
    except Exception as e:
        print(f"Connection glitch: {e}")

# 3. Show the results
print("\n--- Session History ---")
for entry in history:
    print(entry)
```

```python
import matplotlib.pyplot as plt

# 1. Extract the data from our list
x_labels = [f"Check {item['check']}" for item in history]
y_prices = [item['price'] for item in history]

# 2. Create the Plot
plt.figure(figsize=(10, 5))
plt.plot(x_labels, y_prices, marker='o', color='b', linestyle='-', linewidth=2)

# 3. Add the "Story" (Labels)
plt.title(f"CryptoWatch: Real-Time BTC Trend ({history[0]['mood']})")
plt.xlabel("Timeline")
plt.ylabel("Price (USD)")
plt.grid(True)

# 4. Show it!
plt.show()
```

```python
import requests
import time
import matplotlib.pyplot as plt

# 1. INITIALIZATION
history = []
print("🚀 CryptoWatch Live Tracker Started...")

# 2. DATA INGESTION LOOP (Automation)
for i in range(5):
    try:
        # Fetching Price and Sentiment from APIs
        price_res = requests.get("https://api.coinbase.com/v2/prices/BTC-USD/spot")
        btc_price = float(price_res.json()['data']['amount'])
        
        sent_res = requests.get("https://api.alternative.me/fng/")
        mood = sent_res.json()['data'][0]['value_classification']
        
        # Storing data point
        history.append({"check": i+1, "price": btc_price, "mood": mood})
        print(f"Check {i+1}: BTC is ${btc_price} | Sentiment: {mood}")
        
        time.sleep(5) # Wait between checks
    except Exception as e:
        print(f"Error fetching data: {e}")

# 3. ANALYTICS & GENAI LOGIC
start_p = history[0]['price']
end_p = history[-1]['price']
pct_change = ((end_p - start_p) / start_p) * 100

print(f"\n--- AI Analyst Report ---")
print(f"Total Change: {pct_change:.4f}%")

# Heuristic Logic (The "Model")
if pct_change > 0 and "Fear" in history[-1]['mood']:
    print("AI ADVICE: Bullish divergence detected. Buying opportunity despite market fear.")
elif pct_change < 0:
    print("AI ADVICE: Price is sliding. Recommend holding until sentiment improves.")
else:
    print("AI ADVICE: Market is stable. No immediate action required.")

# 4. VISUALIZATION (Storytelling)
plt.figure(figsize=(10, 5))
x = [f"T{item['check']}" for item in history]
y = [item['price'] for item in history]

plt.plot(x, y, marker='o', color='#2ecc71', linewidth=2)
plt.title(f"Bitcoin Price Trend - Current Mood: {history[-1]['mood']}")
plt.ylabel("Price (USD)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
import numpy as np

# 1. Convert our price list into a NumPy Array
prices_array = np.array([item['price'] for item in history])

# 2. Use NumPy to find the average and the spread (Standard Deviation)
average_price = np.mean(prices_array)
volatility = np.std(prices_array)

print(f"--- NumPy Statistics ---")
print(f"Average Price: ${average_price:.2f}")
print(f"Market Volatility: {volatility:.4f}")

if volatility > 5: # If the price swings by more than $5
    print("AI ALERT: High volatility detected. Price is jumping!")
else:
    print("AI ALERT: Low volatility. Price is stable.")
    
```

```python
import requests
import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. DATA INGESTION (Python + APIs)
history = []
print("🚀 Starting Integrated AI Pipeline...")

for i in range(5):
    try:
        price = float(requests.get("https://api.coinbase.com/v2/prices/BTC-USD/spot").json()['data']['amount'])
        mood = requests.get("https://api.alternative.me/fng/").json()['data'][0]['value_classification']
        history.append({"time": i+1, "price": price, "mood": mood})
        print(f"Check {i+1}: ${price}")
        time.sleep(2)
    except:
        print("API connection glitch, skipping...")

# 2. DATA STRUCTURE (Pandas + NumPy)
df = pd.DataFrame(history)
X = df['time'].values.reshape(-1, 1) # Time as Input
y = df['price'].values.reshape(-1, 1) # Price as Target

# 3. PREDICTIVE MODELING (scikit-learn)
# This finds the "Line of Best Fit" to predict the next price
model = LinearRegression().fit(X, y)
next_step = np.array([[6]])
prediction = model.predict(next_step)[0][0]

# 4. HEURISTIC AI SIGNAL (Simulated Neural Logic)
# Instead of PyTorch, we use a weighted logic to get an 'AI Score'
last_price = df['price'].iloc[-1]
ai_signal = (prediction - last_price) / last_price * 100

# 5. GENERATIVE REPORT
print(f"\n--- 🤖 AI ANALYST REPORT ---")
print(f"Current BTC: ${last_price:.2f}")
print(f"ML Prediction (T6): ${prediction:.2f}")

if prediction > last_price:
    advice = "BULLISH: Regression trend suggests an upward move."
else:
    advice = "BEARISH: Trend line is sloping downward. Caution."

print(f"STRATEGY: {advice}")

# 6. VISUALIZATION (Matplotlib)
plt.figure(figsize=(10, 5))
plt.plot(df['time'], df['price'], marker='o', label="Actual Prices")
plt.scatter(6, prediction, color='red', s=100, label="Next-Step Prediction")
plt.title("CryptoWatch: Machine Learning Forecast")
plt.xlabel("Time")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
